```python

list = ["./imagini_auxiliare/" + x for x in os.listdir("./imagini_auxiliare")]
print(list)

game_tables = extract_game(list)
relevant_game_tables = extract_relevant_game(game_tables)

def draw_lines2(images):
    """
    Draw the horizontal and vertical lines onto multiple images.
    :param images:
    :return: Copies of the original images, with lines drawn
    """
    new_images = images.copy()

    lines_vertical,lines_horizontal = get_lines_coords()

    GREEN = (0,255,0)
    RED = (0,0,255)

    for image in tqdm(new_images):
        for line in lines_vertical:
            cv.line(image,line[0],line[1],GREEN,10)

        for line in lines_horizontal:
            cv.line(image,line[0],line[1],RED,10)

    return new_images

img = relevant_game_tables[3]
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
_,table_binary = cv.threshold(img,65,255,cv.THRESH_BINARY_INV)

cv.imwrite("./latex/binary_table.png",table_binary)
```
